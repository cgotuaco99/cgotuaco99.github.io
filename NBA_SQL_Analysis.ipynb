In this project, I'll see if I can identify when NBA players have tended to reach their primes. I will be running SQL commands through pandasql in Python. 

The first step is to download the data and do some exploratory data analysis.

```python
%reset -f 
import pandas as pd
import pandasql as ps
```

Here I load CSV files containing info on individual players and season-long statistics.

```python
Players = pd.read_csv("/Users/christophergotuaco/Desktop/NBAFolder/Players.csv")
SeasonsStats = pd.read_csv("/Users/christophergotuaco/Desktop/NBAFolder/Seasons_Stats.csv")
PlayerData = pd.read_csv("/Users/christophergotuaco/Desktop/NBAFolder/player_data.csv")
```

For this project I'll look at Player Efficiency Rating, or PER. I'll sort from the Season Statistics table by highest to lowest of all time.

```python
ps.sqldf("""SELECT *
            FROM SeasonsStats
            ORDER BY PER DESC
            LIMIT 15;
            """)
```

As formulated, the league average PER will always be 15.00, so it seems there are some obvious outliers here, and many of the names we see are probabaly unfamiliar to even diehard NBA fans.

We can get rid of these outliers by specficing minimums in stats such as games played.
The NBA has a requirement for its league leaders in multiple stats: 70% of team games, or 58 games played in an 82 game season. (https://nba.nbcsports.com/2022/05/19/adam-silver-nba-to-consider-games-played-minimum-for-awards/)

```python
ps.sqldf("""SELECT *
            FROM SeasonsStats
            WHERE G >= 58
            ORDER BY PER DESC
            LIMIT 15;
            """)
```

These players make a bit more sense. Now let's see if we can figure out how to quantify when player's primes begin and end.

We will use the SeasonsStats table to get the max PER year for each unique player, and then take average of the age for each:

```python
newdf = ps.sqldf("""SELECT *
            FROM SeasonsStats
            WHERE G >= 58
            ORDER BY PER DESC;
            """)
```

```python
ps.sqldf("""SELECT Player, MAX(PER), Year, Age
            FROM newdf
            GROUP BY Player
            ORDER BY MAX(PER) DESC
            LIMIT 15;
            """)
```

One issue I came across was the inclusion of the asterisk in players' names, which means that the player was inducted into the Basketball Hall of Fame. Asterisks only appear in the SeasonStats and Player datasets.

```python
ps.sqldf("""SELECT *
            FROM SeasonsStats
            WHERE Player = 'Gary Payton*';
            """)
```

Below we see there are some errors in names: the Gary Payton born in 1992 is actually Gary Payton II. This caused  problems in other cases. One example is two players named David Lee, one who played in 1968, and one who started in 2006.
This caused issues when joining data.

```python
ps.sqldf("""SELECT *
            FROM PlayerData
            WHERE name = 'Gary Payton';
            """)
```

```python
ps.sqldf("""SELECT *
            FROM PlayerData
            WHERE name = 'David Lee';
            """)
```

Let's also take out the asterisk from SeasonStats, which also caused issues in joining data.

```python
Season2 = ps.sqldf("""SELECT REPLACE(Player, '*', '') AS Player_name, *
            FROM SeasonsStats""")
```

To fix the duplicate name problem, we join SeasonStats and PlayerData on the conidition that the player's season year is within their career span.

```python
joindf_total = ps.sqldf("""SELECT *
            FROM Season2 AS s
            LEFT JOIN PlayerData as p
            ON s.Player_name = p.name
            WHERE (s.Year BETWEEN p.year_start AND p.year_end);
            """)
```

Here we make a new column for the span of the player's career.

```python
yearinc = ps.sqldf("""SELECT (Year - year_start), *
            FROM joindf_total;
            """)
```

```python
ps.sqldf("""SELECT name, year, year_start, year_end
            FROM joindf_total
            WHERE name = 'Gary Payton';
            """)
```

Now we will attempt to determine when players reach their primes using PER as our measure.

```python
PER_test = ps.sqldf("""SELECT Player_name, Pos, MAX(PER) AS max_per, Year, Age, year_start, (Year-year_start) AS years_exp
            FROM yearinc
            WHERE G >= 58
            GROUP BY Player
            ORDER BY MAX(PER) DESC
            LIMIT 300;
            """)
```

Now we calculate the average age at which players' peak PERs are reached per position.

```python
ps.sqldf("""SELECT AVG(Age)
            FROM PER_test
            """)
```

```python
ps.sqldf("""SELECT AVG(Age), AVG(years_exp)
            FROM PER_test
            WHERE Pos = 'PG'
            """)
```

```python
ps.sqldf("""SELECT AVG(Age), AVG(years_exp)
            FROM PER_test
            WHERE Pos = 'SG'
            """)
```

```python
ps.sqldf("""SELECT AVG(Age), AVG(years_exp)
            FROM PER_test
            WHERE Pos = 'SF'
            """)
```

```python
ps.sqldf("""SELECT AVG(Age), AVG(years_exp)
            FROM PER_test
            WHERE Pos = 'PF'
            """)
```

```python
ps.sqldf("""SELECT AVG(Age), AVG(years_exp)
            FROM PER_test
            WHERE Pos = 'C'
            """)
```

We see that players who play positions that tend to be taller and heavier peak with slightly fewer years of experience. All players seem to reach their peak PER year around the ages of 26 and 27.
